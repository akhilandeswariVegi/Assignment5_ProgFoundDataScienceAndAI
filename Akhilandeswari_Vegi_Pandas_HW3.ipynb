### Task 17

Alice, Mark, Bill, Tony, Jason, Ravi, Sandeep, Mounika, John and Curtis took three graduate courses `CS 5000`, `CS 5110` and `CS 5600` in Spring 2022. To simulate the midterm scores they got for the courses, please use a random seed 123 to generate a $10 \times 3$ DataFrame called scores to store the 10 stduents' scores. The scores are random integers in the range of `[30, 100]`

```python
np.random.seed(123)
mid_scores = np.random.randint(30, 101, size=(10,3))
students = ['Alice', 'Mark', 'Bill', 'Tony', 'Jason', 'Ravi', 'Sandeep', 'Mounika', 'John', 'Curtis']
courses = ['CS 5000', 'CS 5110', 'CS 5600']

df4 = pd.DataFrame(mid_scores,
                   index = students,
                   columns = courses) 
df4
```

### Task 18

Repeat Task 19 to generate the final scores for those stduents. Please use a random seed of 456 to generate the data

```python
np.random.seed(456)
final_scores = np.random.randint(30,101, size=(10,3))

df5 = pd.DataFrame(final_scores,
                   index = students,
                   columns = courses)
df5
```

### Task 19
Concatenate the two DataFrames from Task 19 and Task 20 into a new DataFrame with multi index

```python
df6 = pd.concat([df4, df5], keys=['Midterm', 'Final'])
df6
```

### Task 20
Name the multi index with `Term` and `Name`, respectively

```python
df6.index.names = ['Term', 'Name']
df6
```

### Task 21
Swap the level of the multi index

```python
df7 = df6.groupby(['Name', 'Term']).first()

#As the final is diplaying first, sorting the indexes [Name and Term] with level and ascending paramteres
df7 = df7.sort_index(level=['Name', 'Term'],
                     ascending=[True, False])

df7
```

### Task 22
Find the average scores for midterm and final for all three courses

```python
df8 = df7.groupby('Term').mean()
df8 = df8.sort_index(level=['Term'],
                     ascending=[False])
df8
```
